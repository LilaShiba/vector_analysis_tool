```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from vectors import Vector
```

```python
df = pd.read_csv('temps.csv', parse_dates=['date'])
```

```python
temps = Vector(label='temps', data_points=df['min'])
plt.plot(temps.v)
```

```python
temps.log_binning()
```

```python
temps.linear_scale()
```

```python
ra =temps.rolling_average(window_size=50)
```
